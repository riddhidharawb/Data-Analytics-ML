```python
import pandas as pd
from sklearn.linear_model import LinearRegression


train_df = pd.read_csv('/content/drive/MyDrive/train.csv')

train_df['TotalSF'] = train_df['1stFlrSF'] + train_df['2ndFlrSF']

X_train = train_df[['TotalSF', 'BedroomAbvGr', 'FullBath']]
y_train = train_df['SalePrice']

test_df = pd.read_csv('/content/drive/MyDrive/test.csv')
test_df['TotalSF'] = test_df['1stFlrSF'] + test_df['2ndFlrSF']


X_test = test_df[['TotalSF', 'BedroomAbvGr', 'FullBath']]


model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)


pd.DataFrame(y_pred, columns=['PredictedPrice']).to_csv('predictions.csv', index=False)

print(pd.DataFrame(y_pred, columns=['PredictedPrice']))

```
